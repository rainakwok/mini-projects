```python
# Import modules 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
```

```python
# Load the dataset from the csv file using pandas
dataset = pd.read_csv('creditcard.csv')
pd.concat([dataset.head(), dataset.tail()])
```

```python
# Find the proportion of fraudulent and non-fraudulent transactions
fraudCount = len(dataset[dataset["Class"] == 1])
validCount = len(dataset[dataset["Class"] == 0])
print("Fraudulent Transactions: " + str(fraudCount))
print("Valid Transactions: " + str(validCount))
print("Proportion of Fraudulent Transactions: " + str(fraudCount/dataset.shape[0]))
```

```python
# Copy the dataset and add a column denoting whether the transaction was fraudulent or genuine
data_prop = dataset.copy()
data_prop[" "] = np.where(data_prop["Class"] == 1 ,  "Fraud", "Genuine")
```

```python
# Plot a pie chart
data_prop[" "].value_counts().plot(
    kind='pie', ylabel=' ', autopct='%1.2f%%',title='Proportion of Fraudulent Transactions')
```
